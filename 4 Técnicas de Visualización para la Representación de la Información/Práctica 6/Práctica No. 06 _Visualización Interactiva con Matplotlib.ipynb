```python
columns_names = ['Dia', 'Paises', 'Casos', 'Muertes']
df = pd.DataFrame(columns=columns_names)
lista =[]
```

```python
for j in range(30):
    dummy='numero'+str(j+1)
    dummy=[]
    for i in range(85):
        dummy.append(j+1)
    dummy2='number'+str(j+1)
    dummy2= pd.DataFrame(dummy, columns=['Dia'])
    
    dummy3='dia'+str(j+1)
    dummy3 = coronapaises[['Paises', 'Casos'+str(j+1), 'Muertes'+str(j+1)]]
    dummy3.columns = ['Paises', 'Casos', 'Muertes']
    
    dummy4='numpais'+str(j+1)
    dummy4=[]
    for i in range(85):
        dummy4.append(i+1)
    dummy5 = 'dfpais'+str(j+1)
    dummy5= pd.DataFrame(dummy4, columns=['NumPais'])
    
    dummy6 = 'result'+str(j+1)
    dummy6= pd.concat([dummy2, dummy3,dummy5], axis=1)
    lista.append(dummy6)
df = pd.concat(lista)
```

```python
df.head()
```

```python
%matplotlib notebook
start_time = 1.00
end_time = 30.00

fig, ax = plt.subplots(1,1)
plt.subplots_adjust(bottom=0.25)
sax = plt.axes([0.25, 0.1, 0.65, 0.03])
slide = Slider(sax, '04/02 al 05/03', start_time, end_time, valinit=start_time)


act_data = df.loc[df['Dia'] == float(int(start_time))]
s, = ax.plot(act_data.NumPais, act_data.Casos, 'o', color='green')
ax.set_title('Casos de Coronavirus en el mundo en 2020')
ax.set_ylabel('NÃºmero de casos')
ax.set_xlabel('PaÃ­ses')

def update(Dia):
    act_data = df.loc[df['Dia'] == float(int(Dia))]
    X = act_data.NumPais
    Y = act_data.Casos

    #Actualizar datos basados en el slider
    s.set_xdata(X)
    s.set_ydata(Y)

    #Reiniciar los limites de los ejes
    ax.set_xlim([X.min()*0.9,X.max()*1.1])
    ax.set_ylim([Y.min()*0.9,Y.max()*1.1])

    fig.canvas.draw()

slide.on_changed(update)
plt.show()
```

```python
%matplotlib notebook
start_time = 1.00
end_time = 30.00

fig, ax = plt.subplots(1,1)
plt.subplots_adjust(bottom=0.25)
sax = plt.axes([0.25, 0.1, 0.65, 0.03])
slide = Slider(sax, 'Del 04/02 al 05/03', start_time, end_time, valinit=start_time)

act_data = df.loc[df['Dia'] == float(int(start_time))]
s, = ax.plot(act_data.NumPais, act_data.Muertes, 'o', color='red')
ax.set_title(' Muertes a casusa del Coronavirus en el mundo 2020')
ax.set_ylabel('NÃºmero de muertes')
ax.set_xlabel('PaÃ­ses')

def update(Dia):
    act_data = df.loc[df['Dia'] == float(int(Dia))]
    X = act_data.NumPais
    Y = act_data.Muertes

    #Actualizar datos basados en el slider
    s.set_xdata(X)
    s.set_ydata(Y)

    #Reiniciar los limites de los ejes
    ax.set_xlim([X.min()*0.9,X.max()*1.1])
    ax.set_ylim([Y.min()*0.9,Y.max()*1.1])

    fig.canvas.draw()


slide.on_changed(update)
plt.show()
```
