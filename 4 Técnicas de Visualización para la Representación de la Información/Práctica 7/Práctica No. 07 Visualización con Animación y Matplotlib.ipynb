```python
from random import sample 
import pandas as pd
import random 
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
```

```python
df = pd.read_csv('C:/Users/HP/Downloads/P7coronaviruscasos.csv')
dfm = pd.read_csv('C:/Users/HP/Downloads/P7coronavirusmuertes.csv')
dfm2 = pd.read_csv('C:/Users/HP/Downloads/P7coronavirusmuertesEST.csv')
df
```

```python
Iran = np.array(df['Iran'])
CoreaDelSur = np.array(df['Corea del Sur'])
Italia = np.array(df['Italia'])
MuertesItalia=np.array(dfm2['Italia'])
MuertesIran=np.array(dfm2['Iran'])
MuertesCorea=np.array(dfm2['Corea del Sur'])
```

```python
Dias=[]
for i in range(38):
    Dias.append(i+1)
```

```python
%matplotlib notebook

fig, ax = plt.subplots()
x1, y1= [],[]
x2, y2= [],[]
x3, y3= [],[]

plt.xlim(0,40)
plt.ylim(0,12500)



sc1 = ax.scatter(x1,y1, c="r", s=MuertesItalia*2000 , alpha=0.6, label='Italia')
sc2 = ax.scatter(x2,y2, c="yellow", s=MuertesIran*2000, alpha=0.6, label='Iran')
sc3 = ax.scatter(x3,y3, c="b", s=MuertesCorea*2000, alpha=0.6, label='CorealDelSur')

def animate(i):
    x1.append(Dias[i])
    y1.append(Italia[i])
    x2.append(Dias[i])
    y2.append(Iran[i])
    x3.append(Dias[i])
    y3.append(CoreaDelSur[i])
    sc1.set_offsets(np.c_[x1,y1])
    sc2.set_offsets(np.c_[x2,y2])
    sc3.set_offsets(np.c_[x3,y3])


    
ax.set_xlabel('Días')
ax.set_ylabel('Casos')
ax.legend(loc='upper left')
ax.set_title('Casos/Muertes COVID-19')

ani =FuncAnimation(fig, animate, frames=37,  interval=100, repeat=False) 
plt.grid() 
plt.show()



```
